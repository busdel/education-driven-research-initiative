```python
# --- 3. BÖLÜM: DÖNGÜLER (LOOPS) ---

olcumler = [110, 125, 130, 115, 140, 85, 98]

print("--- Günlük Şeker Analiz Raporu ---")

# 'for' döngüsü listenin içindeki her bir elemanı 'deger' ismine atar ve sırayla çalıştırır.
for deger in olcumler:
    if deger >= 126:
        # Şeker 126 ve üzeriyse riskli kabul edilir (Örnek mantık)
        status = "YÜKSEK (Risk Grubu)"
    elif deger < 90:
        status = "DÜŞÜK (Kontrol Gerekli)"
    else:
        status = "NORMAL (Sağlıklı)"

    # Her bir ölçüm için sonucu ekrana yazdıralım
    print(f"Ölçüm Değeri: {deger} -> Durum: {status}")

print("--- Analiz Başarıyla Tamamlandı ---")
```

```python
# --- 4. BÖLÜM: FONKSİYONLAR (FUNCTIONS) ---

def analyze_maker(name, sample_list):

    for i in sample_list:
        if i >= 100:
            print(f"{name} has high blood sugar which is {i}")
        elif i <= 50:
            print(f"{name} has low blood sugar which is {i}")
        else:
            print(f"{name} has normal blood sugar which is {i}")

analyze_maker("Onuralp", [140,200,35,60])
```

```python
#5. Bölüm Pandas ile Pratik

import pandas as pd

# 1. DOSYA YÜKLEME (CSV URL üzerinden)
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print(df)

print("--- İlk 5 Satır ---")
print(df.head())
```

```python
# 2. VERİ YAPISINI İNCELEME
print("\n--- Veri Seti Özeti (Sütunlar ve Tipleri) ---")
print(df.info())
```

```python
# 3. MISSING VALUE (Eksik Veri) KONTROLÜ
print("\n--- Eksik Veri Sayıları ---")
# Hangi sütunda kaç tane NaN (boş hücre) var?
print(df.isnull().sum())
```

```python
# 4. TEMEL İSTATİSTİKLER (Mean, Min, Max)
print("\n--- Sayısal Verilerin Özeti ---")
# Bu komut yaş, bilet fiyatı (Fare) gibi sütunların ortalamasını, min/max değerlerini verir
print(df.describe())
```

```python
# Sadece bilet fiyatlarının istatistiksel dağılımı
print(df['Fare'].describe())
```

```python
# 5. TABLO MANTIĞI: FİLTRELEME
# Sadece 1. sınıfta (Pclass=1) seyahat edenlerin yaş ortalamasına bakalım
birinci_sinif_yas = df[df['Pclass'] == 1]['Age'].mean()
print(f"\n1. Sınıf Yolcuların Yaş Ortalaması: {birinci_sinif_yas:.2f}")
```

```python
# 5. TABLO MANTIĞI: GRUPLAMA (Insights)
# "Kadınlar ve çocuklar önce" kuralı işe yaramış mı?
print("\n--- Cinsiyete Göre Hayatta Kalma Oranı ---")
survival_by_sex = df.groupby('Sex')['Survived'].mean()
print(survival_by_sex)
```
