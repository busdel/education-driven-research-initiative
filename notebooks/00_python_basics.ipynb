```python
# --- 2. BÖLÜM: SÖZLÜKLER (DICTIONARIES) ---

# Bir hastanın tüm tıbbi geçmişini tek bir değişkende toplayalım
hasta_dosyasi = {
    "tc_no": "xxxxxxx",
    "ad_soyad": "Patient_1",
    "yas": 30,
    "kan_grubu": "B Rh+",
    "diyabet_durumu": False,
    "son_bes_olcum": [110, 125, 130, 115, 140] # Sözlük içinde liste bile tutabiliriz!
}

# Veriye erişmek: Sadece anahtar ismini yazıyoruz
print(f"Sistemdeki Hasta: {hasta_dosyasi['ad_soyad']}")
print(f"Hastanın Yaşı: {hasta_dosyasi['yas']}")

# Sözlükteki bir veriyi güncelleyelim (Doğum günü gelmiş olsun)
hasta_dosyasi["yas"] = 25

# Yeni bir bilgi ekleyelim (Boy bilgisi sisteme giriliyor)
hasta_dosyasi["boy"] = 170

print("--- Güncel Hasta Kartı ---")
print(hasta_dosyasi)
```

```python
# --- 3. BÖLÜM: DÖNGÜLER (LOOPS) ---

olcumler = [110, 125, 130, 115, 140, 85, 98]

print("--- Günlük Şeker Analiz Raporu ---")

# 'for' döngüsü listenin içindeki her bir elemanı 'deger' ismine atar ve sırayla çalıştırır.
for deger in olcumler:
    if deger >= 126:
        # Şeker 126 ve üzeriyse riskli kabul edilir (Örnek mantık)
        status = "YÜKSEK (Risk Grubu)"
    elif deger < 90:
        status = "DÜŞÜK (Kontrol Gerekli)"
    else:
        status = "NORMAL (Sağlıklı)"

    # Her bir ölçüm için sonucu ekrana yazdıralım
    print(f"Ölçüm Değeri: {deger} -> Durum: {status}")

print("--- Analiz Başarıyla Tamamlandı ---")
```

```python
# --- 4. BÖLÜM: FONKSİYONLAR (FUNCTIONS) ---

def analyze_glucose(name, sample_list):
    results = []

    for i in sample_list:
        if i >= 126:
            results.append(f"{name} has high blood sugar which is {i}")
        elif i < 90:
            results.append(f"{name} has low blood sugar which is {i}")
        else:
            results.append(f"{name} has normal blood sugar which is {i}")

    return results


result = analyze_glucose("Patient_1", [140, 200, 35, 60])
print(result)
```

```python
# 0 = Sağlıklı, 1-4 = Hasta
print("Hastalık Durumu Dağılımı:")
print(df['num'].value_counts())

# Yüzdesel olarak görelim
print("\nOransal Dağılım (%):")
print(df['num'].value_counts(normalize=True) * 100)# normalize  0 ile 1 arasına sıkıştırıyor değerleri
```

```python
# Cinsiyete göre yaş ve kolesterol ortalamaları
print("Cinsiyet Bazlı Ortalamalar:")
display(df.groupby('sex')[['age', 'chol', 'thalch']].mean())
```

```python
# yaşı 50'nin üstü ve kolestrolü 240'tan büyük kaç kişi var bulalım
riskli_grup = df[(df['age'] > 50) & (df['chol'] > 240)]
print(f"Yüksek Riskli Hasta Sayısı: {len(riskli_grup)}")
```

```python
# Kolestrol sütunundaki 0 değerlerini mean value ile dolduralım çünkü 0 diye bir kolestrol ölçüm verisi olamaz

import numpy as np

# 1. Adım: 0 değerlerini NaN (boş) yapalım
df['chol'] = df['chol'].replace(0, np.nan)

# 2. Adım: Tüm sütunun ortalamasını hesaplayalım
genel_ortalama = df['chol'].mean()
print(f"Hesaplanan Genel Ortalama: {genel_ortalama:.2f}")

# 3. Adım: Boş yerleri bu ortalama ile dolduralım
df['chol'] = df['chol'].fillna(genel_ortalama)

# Kontrol:
print(f"Kalan boş değer sayısı: {df['chol'].isnull().sum()}")
```
