```python
# --- Part 1 : Variables and Lists ---

# Simple informations about a patient
patient_name = "Ahmet Zubeyir Kirvar"
patient_age = 20
patient_blood_type = "A+"
patient_weight = 60.5 
patient_height = 176

# List of patient's test results of blood pressure levels
patient_blood_pressure_levels = [120, 130, 125, 135, 130, 100, 110, 90, 85, 140]

# Print patient information
print(f"Name: {patient_name}")
print(f"Age: {patient_age}")
print(f"Blood Type: {patient_blood_type}")
print(f"Weight: {patient_weight} kg")
print(f"Height: {patient_height} cm")
print(f"Blood Pressure Levels: {patient_blood_pressure_levels}")
print(f"First Blood Pressure Level: {patient_blood_pressure_levels[0]}")
```

```python
# --- Part 2 : Dictionaries ---

# Dictionary to store patient's medical history
patient_medical_history = {
    "name" : patient_name,
    "age": patient_age,
    "blood_type": patient_blood_type,
    "height": patient_height,
    "weight": patient_weight,
    "blood pressure levels": patient_blood_pressure_levels,
    "allergies": ["penicillin", "pollen"],
    "medications": ["ibuprofen"],
    "past_illnesses": ["chickenpox"]
}

# Print patient's medical history
print("\nPatient Medical History:")
for key, value in patient_medical_history.items():
    print(f"{key.upper()}: {value}")

# Correcting the patient's weight
patient_medical_history["weight"] = 62.0
print(f"\nUpdated Weight: {patient_medical_history['weight']} kg")

# Adding a new allergy to the patient's medical history
patient_medical_history["allergies"].append("dust")
print(f"Updated Allergies: {patient_medical_history['allergies']}")

# Adding a new information about the patient's vaccination status
patient_medical_history["vaccination_status"] = "up to date"
print(f"Vaccination Status: {patient_medical_history['vaccination_status']}")

# Removing the patient's past illness information
del patient_medical_history["past_illnesses"]

# Printing the new medical history of the patient
print("\nUpdated Patient Medical History:")
for key, value in patient_medical_history.items():
    print(f"{key.upper()}: {value}")
```

```python
# --- Part 3 : Loops and Conditionals ---

# Check if the patient's blood pressure levels are within a normal range
print("\nBlood Pressure Level Analysis:")
for level in patient_blood_pressure_levels:
    if level <= 100:
        status = "Low"
    elif 100 < level < 120:
        status = "Normal"
    elif 120 <= level < 130:
        status = "Elevated"
    elif 130 <= level < 140:
        status = "High Blood Pressure (Stage 1)"
    else:
        status = "High Blood Pressure (Stage 2)"

    print(f"Blood Pressure Level: {level} - Status: {status}")

# Check if the patient has any allergies
print("\nAllergy Check:")
if patient_medical_history["allergies"]:
    print(f"Patient has the following allergies: {', '.join(patient_medical_history['allergies']).upper()}")
```

```python
# --- Part 4 : Functions ---

# Function to calculate the Body Mass Index (BMI) of the patient
def calculate_bmi(weight, height):
    bmi = weight / (height / 100) ** 2
    return bmi

# Calculate and print the patient's BMI
patient_bmi = calculate_bmi(patient_medical_history["weight"], patient_medical_history["height"])
print(f"\nPatient's BMI: {patient_bmi:.2f}")

# Function to determine the BMI category of the patient
def bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obesity"
    
# Determine and print the patient's BMI category
patient_bmi_category = bmi_category(patient_bmi)
print(f"Patient's BMI Category: {patient_bmi_category}")
```

```python
# --- Part 5: Analyzing Patient Data ---

# --- Part 5.1: Using Pandas to Analyze Patient Data ---
# Analyzing different patient informations of diabetic patients

import pandas as pd

df_diabetes = pd.read_csv("diabetes_012_health_indicators_BRFSS2015.csv")


# Display the basic information about the dataset

print("\nDataset Information:")
print(df_diabetes.info())
```

```python
# --- Part 5.2 : Missing Values Analysis ---

# Check for missing values in the dataset
print("\nMissing Values Analysis:")
missing_values = df_diabetes.isnull().sum()
print(missing_values)
```

```python
# --- Part 5.3 : Statistical Summary of the Dataset ---

# Display the statistical summary of the dataset
print("\nStatistical Summary of the Dataset:")
print(df_diabetes.describe())
```

```python
# --- Part 5.4 : Filtering the Dataset for Diabetic Patients ---

# Filter the dataset for diabetic patients
diabetic_patients = df_diabetes[df_diabetes["Diabetes_012"] == 2]
print(f"\nNumber of Diabetic Patients: {len(diabetic_patients)}")
```

```python
# --- Part 6 : Data Visualization ---

import matplotlib.pyplot as plt

# Defining the patient's age distribution among diabetic patients
age_distribution = diabetic_patients["Age"].value_counts().sort_index()

# Plotting the age distribution
plt.figure(figsize=(12, 8))
age_distribution.plot(kind='bar')
plt.title('Age Distribution of Diabetic Patients')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.tight_layout()
plt.show()
```

This chart displays the age distribution of diabetic patients within the dataset. In medical research, it is important to identify the specific age ranges where a disease is most prevalent. This type of visulasion is highly benefecial for public health policymakers when determining which demographic groups should be prioritized for early diagnosis and screening programs.

```python

# Visualizing the relationship between BMI and diabetes status (boxplot)

df_sample = df_diabetes.sample(n=10000, random_state=42)  # Sample 10000 patients for better visualization

plt.figure(figsize=(12, 8))
df_sample.boxplot(column='BMI', by='Diabetes_012')
plt.title('BMI Distribution by Diabetes Status')
plt.xlabel('Diabetes Status')
plt.ylabel('BMI')
plt.tight_layout()
plt.show()
```

This boxplot compares the Body Mass Index (BMI) distribution between non-diabetic (0) and diabetic (1) individuals in the sample. The graph suggests that the median BMI of the diabetic group tends to be higher. Visualizations like this can help researchers observe potential patterns between weight and diabetes, though further statistical analysis is usually needed to confirm any direct correlations.

```python
# Scatter plot to visualize the relationship between age and BMI among diabetic patients

df_sample = diabetic_patients.sample(n=2000, random_state=42)  # Sample 2000 patients for better visualization
plt.figure(figsize=(12, 8))

plt.scatter(df_sample["Age"], df_sample["BMI"], alpha=0.4, color = 'blue')
plt.title('Age vs BMI among Diabetic Patients')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

This scatter plot illustrates the relationship between age and BMI values using a random sample of 1,000 patients. In medical research, scatter plots are widely used to observe how two different risk factors (such as aging and weight gain) interact and cluster within a population. This analysis helps us build a stronger epidemiological profile by revealing whether dangerously high BMI values are concentrated in specific age brackets or distributed across all ages.
